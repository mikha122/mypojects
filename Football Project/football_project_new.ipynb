```python
teams = pd.read_csv('/Users/mikhail/Jupyter/Jupyter-1/Projects/Football Project/teams.csv')
```

```python
data = data.drop(columns=['gameID','teamID_home','teamID_away','ppda_home','ppda_away','goals_home','result_home','goals_away'])
```

```python
def sample_creator(home_team, away_team):
# this for loop will attempt to switch team names into teamID for easier manipulation with the data.    
    for i in range(len(teams)):
        if home_team == teams['name'][i]:
            home_team = teams['teamID'][i]
    for i in range(len(teams)):
        if away_team == teams['name'][i]:
            away_team = teams['teamID'][i]
# this for loop will try to find the previous matches for these teams.
    indexes = []    
    for i in range(len(data)):
        match = data.loc[i]
        if match['awayTeamID'] == away_team and match['homeTeamID'] == home_team:
            indexes.append(i)

# the next loop will attempt to find averages for the home probability of winning, drawing and probability of away winning
    prev_encounters = len(indexes) #this is simply the count of the previous encounters
    
    home_probability = 0
    away_probability = 0
    draw_probability = 0
    for index in indexes:
        game = data.loc[index]
        home_probability += game['homeProbability']
        draw_probability += game['drawProbability']
        away_probability += game['awayProbability']
    home_probability = home_probability / prev_encounters
    away_probability = away_probability / prev_encounters
    draw_probability = draw_probability / prev_encounters

# next, predict the home and away goals at half time
    homeGoalsHalfTime = 0
    awayGoalsHalfTime = 0
    for index in indexes:
        game = data.loc[index]
        homeGoalsHalfTime += game['homeGoalsHalfTime']
        awayGoalsHalfTime += game['awayGoalsHalfTime']
    homeGoalsHalfTime = homeGoalsHalfTime / prev_encounters
    awayGoalsHalfTime = awayGoalsHalfTime / prev_encounters
    
# next, i will find the xGoals_home, shots_home, shotsOnTarget, deep_home, fouls_home, corners_home, yellow and red cards
    xGoals_home = 0
    shots_home = 0
    shotsOnTarget_home = 0
    deep_home = 0
    fouls_home = 0
    corners_home = 0
    yellowCards_home = 0
    redCards_home = 0
    for index in indexes:
        game = data.loc[index]
        xGoals_home += game['xGoals_home']
        shots_home += game['shots_home']
        shotsOnTarget_home += game['shotsOnTarget_home']
        deep_home += game['deep_home']
        fouls_home += game['fouls_home']
        corners_home += game['corners_home']
        yellowCards_home += game['yellowCards_home'] 
        redCards_home += game['redCards_home']
    
    xGoals_home = xGoals_home / prev_encounters
    shots_home = shots_home / prev_encounters
    shotsOnTarget_home = shotsOnTarget_home / prev_encounters
    deep_home = deep_home / prev_encounters
    fouls_home = fouls_home / prev_encounters
    corners_home = corners_home / prev_encounters
    yellowCards_home = yellowCards_home / prev_encounters
    redCards_home = redCards_home / prev_encounters


#Finally, i will do the same for the away team. That will conclude the data calculations. 
    xGoals_away = 0
    shots_away = 0
    shotsOnTarget_away = 0
    deep_away = 0
    fouls_away = 0
    corners_away = 0
    yellowCards_away = 0
    redCards_away = 0
    for index in indexes:
        game = data.loc[index]
        xGoals_away += game['xGoals_away']
        shots_away += game['shots_away']
        shotsOnTarget_away += game['shotsOnTarget_away']
        deep_away += game['deep_away']
        fouls_away += game['fouls_away']
        corners_away += game['corners_away']
        yellowCards_away += game['yellowCards_away'] 
        redCards_away += game['redCards_away']
    
    xGoals_away = xGoals_away / prev_encounters
    shots_away = shots_away / prev_encounters
    shotsOnTarget_away = shotsOnTarget_away / prev_encounters
    deep_away = deep_away / prev_encounters
    fouls_away = fouls_away / prev_encounters
    corners_away = corners_away / prev_encounters
    yellowCards_away = yellowCards_away / prev_encounters
    redCards_away = redCards_away / prev_encounters


# then we need to output the data in the right format
    result = pd.DataFrame([[home_team,away_team, home_probability, draw_probability, away_probability, homeGoalsHalfTime,awayGoalsHalfTime,xGoals_home,shots_home,shotsOnTarget_home,deep_home,fouls_home,corners_home,yellowCards_home,redCards_home,xGoals_away,shots_away,shotsOnTarget_away,deep_away,fouls_away,corners_away,yellowCards_away,redCards_away]],columns=data.columns)
    return result
```

```python
WestHam_Watford = sample_creator('West Ham','Watford')
```

```python
WestHam_Watford.to_csv('Leeds_Newcastle.csv')
```
