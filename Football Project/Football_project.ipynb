```python
new_teamstats = pd.DataFrame(columns=teamstats.columns)
```

```python
new_teamstats = teamstats
```

```python
#creates 2 new datasets which now have the columns for the home and away games in order to append horizontally to each other
teamstats_home = pd.DataFrame(columns = new_teamstats.columns)
teamstats_away = pd.DataFrame(columns = new_teamstats.columns)
```

```python
#this separates the home and away matches to later merge them together
for row in range(1, len(new_teamstats)):
    if new_teamstats.iloc[row]['gameID'] == new_teamstats.iloc[row-1]['gameID']:
        teamstats_away = teamstats_away.append(new_teamstats.iloc[row])
        teamstats_home = teamstats_home.append(new_teamstats.iloc[row-1])
```

```python
merged_teamstats = teamstats_home.merge(teamstats_away, left_on='gameID',right_on='gameID',suffixes=('_home','_away'))
```

```python
merged_teamstats
```

```python
merged_gameresult = merged_teamstats['result_home']
```

```python
#this encoder will change win letters to numbers, meaning that it will be easier to interpret them for the computer. W - 2, D - 1, L - 0 (for home team)
number_clean = { 'W':2,
                'L':0,
                'D':1
}
merged_gameresult = merged_gameresult.replace(number_clean)
merged_gameresult.head()
```

```python
#this deletes the current result home column and adds the one that puts number there instead of letters
merged_teamstats['result_home'] = merged_gameresult
```

```python
merged_teamstats = merged_teamstats.drop(columns=['location_away','location_home','result_away'])
```

```python
games_data = games_data.drop(columns=['homeGoals','awayGoals'])
```

```python
#this is the fully merged dataset, final data needed. 
all_merged = games_data.merge(merged_teamstats,left_on='gameID',right_on='gameID')
```
